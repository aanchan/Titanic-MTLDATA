Pandas and Scikit-Learn for the Titanic Challenge
---
There 5 broad steps that are covered in this Notebook:
1. Loading the Titanic CSV data into a Pandas Dataframe
2. Use Pandas for 'Data Munging' and 'Feature Engineering'
    i.e. Format + clean the training and test data to feed into Scikit-learn for machine-learning.
3. Scikit-learn for machine learning is incompatible with Pandas,so convert the data from a Pandas DataFrame into a Numpy array.
4. Load the resulting training data and supervision labels to train or fit a model using Scikit-learn.
5. See how well your model did by predicting outcomes on the test data.

```python
df['Gender'] = 4
df.head()
```

```python
df['Gender'] = df['Sex'].map( lambda x: x[0].upper() )
df.head()
```

```python
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df.head()
```

```python
df['Boarding'] = 4
df.head()
```

```python
df['Boarding'] = df['Embarked'].map( {'S': 0, 'C': 1, 'Q':2} ).astype(int)
df.head()
```

```python
df[df['Embarked'].isnull()]
```

```python
df['Boarding'] = df[df['Embarked'].notnull()]['Embarked'].map( {'S': 0, 'C': 1, 'Q':2} ).astype(int)
df.head()
```

```python
df[df['Embarked'].isnull()]['Boarding']
```

```python
df[df['Embarked'].isnull()]
```

```python
median_ages = np.zeros((2,3))#Rows correspond to females, males, columns correspond to the class
median_ages
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df[(df['Gender'] == i) & \
                              (df['Pclass'] == j+1)]['Age'].dropna().median()
 
median_ages
```

```python
df['AgeFill'] = df['Age']

df.head()
```

```python
df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        df.loc[ (df.Age.isnull()) & (df.Gender == i) & (df.Pclass == j+1),\
                'AgeFill'] = median_ages[i,j]
df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFill']].head(10)
```

```python
df['AgeIsNull'] = pd.isnull(df.Age).astype(int)
```

```python
df.describe()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
df['FamilySize'].head(10)
```

```python
df['Age*Class'] = df.AgeFill * df.Pclass
df['Age*Class'].head()
```

```python
df.dtypes
```

```python
df.dtypes[df.dtypes.map(lambda x: x=='object')]
```

```python
df = df.drop(['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked','Age','Boarding'], axis=1) 
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
train_targets = df['Survived'].values
train_targets
df=df.drop(['Survived'],axis=1)
train_data = df.values
train_data
```

```python

clf = LogisticRegression()
```

```python
logReg = clf.fit(train_data,train_targets)
```
